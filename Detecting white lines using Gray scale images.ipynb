```python
#import all libraries
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#import the rgb image and convert to grayscale
image_color=mpimg.imread('grayroad.jfif')
image_gray=cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap='gray')
```

```python
#check size of image
sz_gray=image_gray.shape
sz_gray
```

```python
#apply filter to check white pixels of image
image_copy=np.copy(image_gray)

#any value not white color
image_white=image_copy[(image_gray[:,:]<250)]=0
image_white
```
