```python
import cv2
import numpy as np
import matplotlib.image as mpimg
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#import image
image=cv2.imread("road.jpg")
cv2.imshow('test image',image)
cv2.waitKey(0)
cv2.destroyAllWindows
```

```python
plt.imshow(image)
```

```python
#Apply Canny edge detection
image_2=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
image_canny=cv2.Canny(image_2,50,200,apertureSize=3)
image_canny
```

```python
cv2.imshow('canny image',image_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Apply Hough transform
lines=cv2.HoughLines(image_canny,1,np.pi/180,300)
if lines is not None:
    for i in range(0,len(lines)):
        rho=lines[i][0][0]
        theta=lines[i][0][1]
    
```

```python
x0=rho*np.cos(theta)
y0=rho*np.sin(theta)
```

```python
a=rho*np.cos(theta)
b=rho*np.sin(theta)
```

```python
x1=int(x0+1000*(-b))
y1=int(y0+1000*(a))
x2=int(x0-1000*(-b))
y2=int(y0-1000*(a))
```

```python
cv2.line(image,(x1,y1),(x2,y2),(255,0,0),2)
```

```python
cv2.imshow('Hough Lines',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
