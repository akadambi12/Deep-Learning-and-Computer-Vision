```python
#import libraries-
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image_color=cv2.imread('road.jpg')
cv2.imshow('test image',image_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
plt.imshow(image_color)#does not show accurate colors.Its all in BGR
```

```python
height,width=image_color.shape[:2]
height
```

```python
width
```

```python
#convert to gray image
image_gray=cv2.cvtColor(image_color,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap='gray')
```

# masking regions of interest

```python

#1. select points of region of ineterest.
ROI=np.array([[(0,2100),(2000,750),(2500,1500),(3500,height)]],dtype=np.int32)
```

```python
#2. define blank image with all zeros
blank=np.zeros_like(image_gray)
blank.shape
```

```python
#3. fill the region of interest in white
mask=cv2.fillPoly(blank,ROI,255)
```

```python
#4. perform bitwise and operation to select only region of interest
masked_image=cv2.bitwise_and(image_gray,mask)
plt.imshow(masked_image,cmap='gray')
```
