```python
import cv2
import numpy as np
import matplotlib.image as mpimg
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#import the rgb image
image_color=mpimg.imread('Road-markings.webp')
plt.imshow(image_color)
image=cv2.imread('Road-markings.webp')
cv2.imshow('test image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#convert rgb image to grayscale image
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow('gray image',gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#smoothing the image
blur_conv=cv2.GaussianBlur(gray,(5,5),0)
cv2.imshow('blurred image',blur_conv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Canny Edge detection
canny_conv=cv2.Canny(blur_conv,50,155)
cv2.imshow('Canny edge detection',canny_conv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Masking region of interest
def reg_of_interest(image):
    mask_region=np.array([[(100,image.shape[0]),(550,image.shape[0]),(350,50)]])
    image_masked=np.zeros_like(image)
    cv2.fillPoly(image_masked,mask_region,255)
    return image_masked
```

```python
cv2.imshow('Result',reg_of_interest(canny_conv))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
